```python
import pandas as pd
```

```python
df = pd.read_csv('crime-data_crime-data_crimestat.csv')
```

```python

print(df.columns)


df.isnull().sum().sort_values(ascending=False).head(10)
```

```python
sorted(df['ZIP'].dropna().unique())
```

```python
crime_by_zip = df.groupby('ZIP').size().sort_values(ascending=False)

print("Top 10 ZIP codes by total incidents:")
print(crime_by_zip.head(10))
```

```python
crime_by_type = df.groupby('UCR CRIME CATEGORY').size().sort_values(ascending=False)

print("Most common crime types:")
print(crime_by_type.head(10))
```

```python
df['OCCURRED_ON_DATE'] = pd.to_datetime(df['OCCURRED ON'], errors='coerce')
df['YEAR'] = df['OCCURRED_ON_DATE'].dt.year

crime_trend = df.groupby('YEAR').size()
crime_trend.plot(kind='line', marker='o', title='Total Crimes per Year', figsize=(8,4))
```

```python
if 'VIOLENT_CRIME_FLAG' in df.columns:
    violent = df[df['VIOLENT_CRIME_FLAG'] == True]
    nonviolent = df[df['VIOLENT_CRIME_FLAG'] == False]

    print("Top ZIPs for Violent Crime:")
    print(violent.groupby('ZIP_CODE')['INCIDENT_ID'].count().sort_values(ascending=False).head(10))
else:
    print("No 'VIOLENT_CRIME_FLAG' column found  skipping this section.")
```

```python
print("\n===== SUMMARY REPORT =====")
print(f"Total rows: {len(df)}")
print(f"Total ZIPs: {df['ZIP'].nunique()}")
print("\nTop 5 ZIP codes by total incidents:")
print(crime_by_zip.head(5))
print("\nTop 5 crime types:")
print(crime_by_type.head(5))
print("\nYear-to-year total crime counts:")
print(crime_trend)
```

```python
crime_by_zip.to_csv('crime_by_zip_summary.csv')
crime_by_type.to_csv('crime_by_type_summary.csv')

print(" Summaries saved to CSV files in your Colab workspace.")
```

```python
import matplotlib.pyplot as plt

violent = ['HOMICIDE', 'AGGRAVATED ASSAULT', 'ROBBERY', 'SEXUAL ASSAULT']

df['CRIME TYPE'] = df['UCR CRIME CATEGORY'].apply(lambda x: 'Violent' if any(v in str(x).upper() for v in violent) else 'Non-Violent')

df['CRIME TYPE'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title('Violent vs Non-Violent Crimes')
plt.ylabel('')
plt.show()
```

```python
crime_zip_year = df.groupby(['ZIP', 'YEAR']).size().unstack()
crime_growth = crime_zip_year.pct_change(axis=1).mean(axis=1).sort_values(ascending=False)
crime_growth.head(10)
```
